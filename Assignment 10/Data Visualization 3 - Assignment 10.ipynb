```python
q1 = df.sepal_width.quantile(0.25)
q3 = df.sepal_width.quantile(0.75)
q1,q3
```

```python
iqr = q3-q1
iqr
```

```python
l_limit = q1-1.5*iqr
u_limit = q3+1.5*iqr
l_limit,u_limit
```

```python
df[(df['sepal_width']<l_limit) | (df['sepal_width']>u_limit)]


```

```python
df =df[(df['sepal_width']>l_limit) & (df['sepal_width']<u_limit)]
```
