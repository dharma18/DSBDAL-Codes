```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['species'] = le.fit_transform(df['species'])
```

```python
df.head(7)
```

```python
x = df.drop(labels='species',axis =1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
xtr_scale = sc.fit_transform(x_train)
xte_scale = sc.transform(x_test)
```

```python
xtr_scale
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(xtr_scale,y_train)
```

```python
y_pred = gnb.predict(xte_scale)
y_pred
```

```python
pred_df=pd.DataFrame(np.c_[ y_test, y_pred], columns = ["original","predicted"])
pred_df
```

```python
from sklearn.metrics import accuracy_score
gnb.score(xte_scale,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report as cr

print(cr(y_test,y_pred))
```
